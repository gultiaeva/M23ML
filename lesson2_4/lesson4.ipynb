### Локальная регрессия

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
y_func = (-3 + 2*x*x)
y = y_func + 40 * np.random.randn(*x.shape)
```

```python
plt.scatter(x, y, marker='.')
```

```python
from localreg import *
```

```python
y0 = localreg(x, y, degree=0, kernel=rbf.tricube, radius=1)
y1 = localreg(x, y, degree=1, kernel=rbf.tricube, radius=1)
y2 = localreg(x, y, degree=2, kernel=rbf.tricube, radius=1)
```

```python
plt.plot(x, y, '.', markersize=0.8, color='gray')
plt.plot(x, y_func, label='Ground truth')
plt.plot(x, y0, label='Moving average')
plt.plot(x, y1, label='Local linear regression')
plt.plot(x, y2, label='Local quadratic regression')
plt.legend()
plt.show()
```
